```python
def run():
    lr = 0.018
    num_itr = 4000
    m = gd(lr,num_itr)
    print(m)
    return (m)
```

```python
def gd(lr,num_itr):
    size = x_train.shape
    M = size[0]
    N = size[1]
    m = np.zeros(N)
    for i in range(num_itr):
        m = step_gradient(lr,m)
        print(i,cost(m))
    return m
```

```python
def step_gradient(lr,m):
    size = x_train.shape
    M = size[0]
    N = size[1]
    m_slope = np.zeros(N)
    for i in range(M):
        y = y_train[i]
        x = x_train[i]
        for j in range(N):
            m_slope[j] += (-2/M) * (y - ((m*x).sum())) * x[j]
    for i in range(N):
        m[i] = m[i] - lr*m_slope[i]
    return m
```

```python
def cost(m):
    size = x_train.shape
    M = size[0]
    N = size[1]
    cost = 0
    for i in range(M):
        y = y_train[i]
        x = x_train[i]
        cost += (1/M) * (y - (m*x).sum())**2
    return cost
```

```python
def predict(x,m):
    size = len(x)
    y_pred = np.zeros(size)
    for i in range(size):
        y_pred[i]= (m*x[i]).sum()
    return y_pred
```

```python
def score(y_pred,y_truth):
    u = ((y_truth - y_pred)**2).sum()
    v = ((y_truth - y_truth.mean())**2).sum()
    return 1-(u/v)
```
